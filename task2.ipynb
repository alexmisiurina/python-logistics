```python
users = []
chats = []
messages = []

def add_user(username, created_at):
    newuser = {
        "id": len(users),
        "username": username,
        "created_at": created_at
    }
    users.append(newuser)
    return newuser["id"]

def new_chat(users, name, created_at):
    newchat = {
        "id": len(chats),
        "name": name,
        "users": users,
        "created_at": created_at
    }
    chats.append(newchat)
    return newchat["id"]

def send_message(user, chat, message, created_at):
    newmessage = {
        "id": len(messages),
        "chat": chat,
        "author": user,
        "text": message,
        "created_at": created_at
    }
    messages.append(newmessage)
    return newmessage["id"]

def get_messages(chat):
    list = []
    for a in messages:
        if chat == a["chat"]:
            list.append(a)

    list.sort(key=lambda a: a["created_at"])

    return list

def get_chats(user):
    list = []
    for a in chats:
        if user in a["users"]:
            list.append((a, get_messages(a)[-1]))
    
    list.sort(key=lambda a: a[1])
    newlist = []
    for a in list:
        newlist.append(a[0])

    return newlist
```
