```python
from bs4 import BeautifulSoup
import requests

url = "https://adme.media/svoboda-psihologiya/100-voprosov-kotorye-nuzhno-obsudit-s-rebenkom-chtoby-ukrepit-s-nim-svyaz-2043665/"

response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')
asks = soup.find_all('div', class_="e599d42b92530cddc130")

dict = {}

current = ""
for ask in asks:
    if '<div class="e599d42b92530cddc130"><h3 style="text-align: center;">' in str(ask):
        current = ask.text
    else:
        dict[current] = ask.text.replace("\xa0", " ").replace("&nbsp", " ").split("\n")

print(dict)
```
